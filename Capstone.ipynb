Historical Method

```python
import pandas as pd
import numpy as np

def calculate_var_historical(file_path, confidence_level=0.95):
    # Load Excel file
    df = pd.read_excel(file_path)

    # Ensure the correct column is used (assuming column name is 'PORT_RET')
    if 'PORT_RET' not in df.columns:
        raise ValueError("The Excel file must contain a 'PORT_RET' column.")

    # Sort returns in ascending order
    sorted_returns = df['PORT_RET'].sort_values().reset_index(drop=True)

    # Determine the percentile index
    var_index = int((1 - confidence_level) * len(sorted_returns))

    # Get the VaR value
    var_value = sorted_returns.iloc[var_index]

    return var_value

# Example usage
file_path = "capstone1.xlsx"  # Update with your actual file path
confidence_level = 0.95  # 95% confidence level
var = calculate_var_historical(file_path, confidence_level)
print(f"Value at Risk (VaR) at {confidence_level*100}% confidence level: {var}")
```

Covarience VAR Method

```python
import pandas as pd
import numpy as np
import scipy.stats as st

def calculate_var_covariance(file_path, confidence_level=0.95):
    # Load Excel file
    df = pd.read_excel(file_path)

    # Ensure the correct column is used (assuming column name is 'PORT_RET')
    if 'PORT_RET' not in df.columns:
        raise ValueError("The Excel file must contain a 'PORT_RET' column.")

    # Calculate mean and standard deviation of PORT_RET
    mean_return = df['PORT_RET'].mean()
    std_dev = df['PORT_RET'].std()

    # Get the z-score for the given confidence level
    z_score = st.norm.ppf(1 - confidence_level)

    # Calculate VaR using the covariance method
    var_value = mean_return + z_score * std_dev

    return var_value

# Example usage
file_path = "capstone1.xlsx"  # Update with your actual file path
confidence_level = 0.95  # 95% confidence level
var = calculate_var_covariance(file_path, confidence_level)
print(f"Value at Risk (VaR) at {confidence_level*100}% confidence level: {var}")
```

Monte Carlo Method

```python
def calculate_var_monte_carlo(file_path, confidence_level=0.95, simulations=10000):
    # Load Excel file
    df = pd.read_excel(file_path)

    # Ensure the correct column is used (assuming column name is 'PORT_RET')
    if 'PORT_RET' not in df.columns:
        raise ValueError("The Excel file must contain a 'PORT_RET' column.")

    # Calculate mean and standard deviation of PORT_RET
    mean_return = df['PORT_RET'].mean()
    std_dev = df['PORT_RET'].std()

    # Simulate returns using normal distribution
    simulated_returns = np.random.normal(mean_return, std_dev, simulations)

    # Determine the percentile corresponding to the confidence level
    var_value = np.percentile(simulated_returns, (1 - confidence_level) * 100)

    return var_value

# Example usage
file_path = "capstone1.xlsx"  # Update with your actual file path
confidence_level = 0.95  # 95% confidence level

var_monte_carlo = calculate_var_monte_carlo(file_path, confidence_level)
print(f"Monte Carlo VaR at {confidence_level*100}% confidence level: {var_monte_carlo}")
```

```python
import scipy.stats as st
import tensorflow as tf
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout
from sklearn.preprocessing import MinMaxScaler

def calculate_var_lstm(file_path, confidence_level=0.95, time_steps=10, epochs=10):
    df = pd.read_excel(file_path)
    if 'PORT_RET' not in df.columns:
        raise ValueError("The Excel file must contain a 'PORT_RET' column.")

    data = df[['PORT_RET']].values
    scaler = MinMaxScaler()
    data_scaled = scaler.fit_transform(data)

    X, y = [], []
    for i in range(len(data_scaled) - time_steps):
        X.append(data_scaled[i:i+time_steps])
        y.append(data_scaled[i+time_steps])
    X, y = np.array(X), np.array(y)

    model = Sequential([
        LSTM(50, return_sequences=True, input_shape=(time_steps, 1)),
        Dropout(0.2),
        LSTM(50),
        Dropout(0.2),
        Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')

    model.fit(X, y, epochs=epochs, batch_size=16, verbose=0, validation_split=0.1)

    X_test = data_scaled[-time_steps:].reshape(1, time_steps, 1)
    predicted_return_scaled = model.predict(X_test)
    predicted_return = scaler.inverse_transform(predicted_return_scaled.reshape(-1, 1))

    return np.percentile(predicted_return, (1 - confidence_level) * 100)

file_path = "capstone1.xlsx"
confidence_level = 0.95

var_lstm = calculate_var_lstm(file_path, confidence_level)
print(f"LSTM VaR at {confidence_level*100}% confidence level: {var_lstm}")
```

```python
# Single step dataset preparation
def singleStepSampler(df, window):
    xRes = []
    yRes = []
    for i in range(0, len(df) - window):
        res = []
        for j in range(0, window):
            r = []
            for col in df.columns:
                r.append(df[col][i + j])
            res.append(r)
        xRes.append(res)
        yRes.append(df[['Open', 'Close']].iloc[i + window].values)
    return np.array(xRes), np.array(yRes)
```

```python
# Dataset splitting
SPLIT = 0.85
(xVal, yVal) = singleStepSampler(df_scaled, 20)
X_train = xVal[:int(SPLIT * len(xVal))]
y_train = yVal[:int(SPLIT * len(yVal))]
X_test = xVal[int(SPLIT * len(xVal)):]
y_test = yVal[int(SPLIT * len(yVal)):]
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense
from tensorflow.keras.optimizers import Adam

# Load the data
def load_portfolio_returns(filename):
    """
    Load portfolio returns from Excel file

    Args:
        filename (str): Path to the Excel file

    Returns:
        pandas.DataFrame: DataFrame with portfolio returns
    """
    df = pd.read_excel(filename)
    df['Date'] = pd.to_datetime(df['Date'])
    df.set_index('Date', inplace=True)
    return df

# Prepare data for LSTM
def prepare_lstm_data(returns, look_back=10):
    """
    Prepare data for LSTM model

    Args:
        returns (numpy.array): Portfolio returns
        look_back (int): Number of previous time steps to use

    Returns:
        tuple: Reshaped X and y data for LSTM
    """
    X, y = [], []
    for i in range(len(returns) - look_back):
        X.append(returns[i:i+look_back])
        y.append(returns[i+look_back])
    return np.array(X), np.array(y)

# Build LSTM Model
def create_lstm_model(look_back):
    """
    Create LSTM model for return prediction

    Args:
        look_back (int): Number of previous time steps

    Returns:
        tensorflow.keras.Model: Compiled LSTM model
    """
    model = Sequential([
        LSTM(50, activation='relu', input_shape=(look_back, 1), return_sequences=True),
        LSTM(50, activation='relu'),
        Dense(1)
    ])
    model.compile(optimizer=Adam(learning_rate=0.001), loss='mse')
    return model

# Calculate VaR using LSTM predictions
def calculate_var(model, scaler, returns, confidence_level=0.95, look_back=10):
    """
    Calculate Value at Risk (VaR) using LSTM model

    Args:
        model (tensorflow.keras.Model): Trained LSTM model
        scaler (MinMaxScaler): Scaler used for data normalization
        returns (numpy.array): Portfolio returns
        confidence_level (float): Confidence level for VaR
        look_back (int): Number of previous time steps

    Returns:
        float: Value at Risk (VaR)
    """
    # Prepare input data
    scaled_returns = scaler.transform(returns.reshape(-1, 1))

    # Create input sequence
    input_seq = scaled_returns[-look_back:].reshape(1, look_back, 1)

    # Generate predictions
    predictions = []
    num_simulations = 100

    for _ in range(num_simulations):
        # Predict next return
        predicted_return = model.predict(input_seq)

        # Append prediction and update input sequence
        predictions.append(predicted_return[0, 0])
        input_seq = np.roll(input_seq, -1, axis=1)
        input_seq[0, -1, 0] = predicted_return[0, 0]

    # Inverse transform predictions
    predictions = scaler.inverse_transform(np.array(predictions).reshape(-1, 1))

    # Calculate VaR
    var = np.percentile(predictions, (1 - confidence_level) * 100)

    return var

def main():
    # Load data
    df = load_portfolio_returns('capstone1.xlsx')
    returns = df['PORT_RET'].values

    # Normalize data
    scaler = MinMaxScaler()
    scaled_returns = scaler.fit_transform(returns.reshape(-1, 1))

    # Set hyperparameters
    look_back = 11

    # Prepare data
    X, y = prepare_lstm_data(scaled_returns, look_back)
    X = X.reshape(X.shape[0], look_back, 1)

    # Create and train LSTM model
    model = create_lstm_model(look_back)
    model.fit(X, y, epochs=20, batch_size=32, verbose=0)

    # Calculate VaR
    var_95 = calculate_var(model, scaler, returns, confidence_level=0.95)
    var_99 = calculate_var(model, scaler, returns, confidence_level=0.99)

    print(f"95% Value at Risk (VaR): {var_95:.4f}")
    print(f"99% Value at Risk (VaR): {var_99:.4f}")

    # Visualize results
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, returns, label='Portfolio Returns')
    plt.title('Portfolio Returns with VaR Levels')
    plt.xlabel('Date')
    plt.ylabel('Log Returns')
    plt.axhline(y=var_95, color='r', linestyle='--', label='95% VaR')
    #plt.axhline(y=var_99, color='g', linestyle='--', label='99% VaR')
    plt.legend()
    plt.show()

if __name__ == "__main__":
    main()
```
